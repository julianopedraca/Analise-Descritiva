## Distribuição da Avaliação por Rating

```python
data['Avaliação']  = None #criando a coluna avaliação
```

```python
#atribuindo valor da avaliação através da coluna user_rating_score
data['Avaliação'] = list(map(lambda x: 'Ótimo' if (x > 80 and x <=100) else \
                                       ('Bom' if (x > 60 and x <= 80) else \
                                       ('Ok' if (x > 40 and x <= 60) else \
                                       ('Ruim' if (x > 20 and x <= 40) else 'Péssimo'))), data['user_rating_score']))
```

```python
data.head(60)
```

```python
round(data.groupby(by=['Avaliação','rating'])['user_rating_score'].mean(),ndigits=2)
```

### Filmes Ótimos por Rating

```python
filmes_otimos = data.query("Avaliação == 'Ótimo'")
```

```python
round(filmes_otimos.groupby(by=['Avaliação','rating'])['user_rating_score'].mean().sort_values(ascending=False),ndigits= 2)
```

### Filmes Ok por Rating

```python
filmes_ok = data.query("Avaliação == 'Ok'")
```

```python
round(filmes_ok.groupby(by=['Avaliação','rating'])['user_rating_score'].mean().sort_values(ascending=False),ndigits=2)
```

### Filmes Bons por Rating

```python
filmes_bom = data.query("Avaliação =='Bom'")
```

```python
round(filmes_bom.groupby(by=['Avaliação','rating'])['user_rating_score'].mean().sort_values(ascending=False),ndigits=2)
```

### Distribuição de Frequência Avaliação x Rating
